```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer #for each movie
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("movies1.csv")
df
```

```python
df.columns
```

```python
df.shape
```

```python
features=['genres','keywords','original_language','title','cast','director']
for feature in features:
    df[feature]=df[feature].fillna('') #filling missing values wuth space
```

```python
def combined_features(row):
    return row['title']+","+row['genres']+","+row['keywords']+","+row['original_language']+","+row['cast']+","+row['director']
df['combined_features'] = df.apply(combined_features,axis=1) #column wise axis=1 dataframe same columns merge
df['combined_features'] #one more column with all those combined features
```

```python
tfid=TfidfVectorizer()
tfidv=tfid.fit_transform(df['combined_features']) #text word vector
tfidv.toarray()
```

```python
tfidv.shape
```

```python
cosine_sim=cosine_similarity(tfidv)
cosine_sim #compare the vectors
```

```python
cosine_sim.shape
```

```python
movie=input("Enter movie name: ")
def get_index(mn):
    return df[df.title==mn].index[0]
mi=get_index(movie)
mi
```

```python
sm=list(enumerate(cosine_sim[mi])) #0 se aage badhenge
print(sm)
```

```python
sorted_sm=sorted(sm,key=lambda x:x[1],reverse=True) #sort ACCORDING TO COSINE SIMILARITY AND NOT INDEX (for similar movies)
print(sorted_sm)
```

```python
def get_info(index):
    return df[df.index==index]['title'].values[0]+": "+ df[df.index==index]['cast'].values[0]
i=0
for movie in sorted_sm:
    print(get_info(movie[0]))
    i=i+1 #FOR NEXT RECOMMENDATION
    if i>10:
        break
```
